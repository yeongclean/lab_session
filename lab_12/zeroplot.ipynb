```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# CSV 파일을 읽을 때, header가 없으므로 header=None 옵션을 사용하고,
# 각 칼럼에 material_id, exp_bandgap, pred_bandgapd 이름을 할당
data = pd.read_csv('test_results_9.csv', header=None, names=['material_id', 'exp_bandgap', 'pred_bandgapd'])

# Plot scatter
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='exp_bandgap', y='pred_bandgapd')
plt.plot([data.exp_bandgap.min(), data.exp_bandgap.max()],
         [data.exp_bandgap.min(), data.exp_bandgap.max()], 'r--')  # y=x line
plt.xlabel('calculated Band Gap')
plt.ylabel('Predicted Band Gap')
plt.title('Comparison of Experimental and Predicted Band Gaps')
plt.show()


```
